```python
colors = ['white', 'green', 'yellow', 'red', 'brown', 'pink']
color_rdd = sc.parallelize(colors)
keyval_rdd = color_rdd.map(lambda x:(x, len(x)))
color_df = keyval_rdd.toDF(['color', 'length']) 
```

```python
import pyspark.sql.functions as F
from pyspark.sql import Window

(
  color_df
  .withColumn(
    'min_length',
    F.min('length').over(Window.rowsBetween(Window.unboundedPreceding, Window.unboundedFollowing))
  )
  .withColumn(
    'range_length',
    F.max('length').over(Window.rowsBetween(Window.unboundedPreceding, Window.unboundedFollowing))
    - F.col('min_length')
  )
  .withColumn(
    'max_min_normalized_length',
    (F.col('length') - F.col('min_length')) / F.col('range_length')
  )
  .select('color', 'length', 'max_min_normalized_length')
  .show()
)
```
