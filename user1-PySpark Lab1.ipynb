```python
colors = ['white','green','yellow','red','brown','pink']
```

```python
color_rdd = sc.parallelize(colors)
```

```python
keyval_rdd = color_rdd.map(lambda x:(x,len(x)))
```

```python
color_df = keyval_rdd.toDF(['color','length'])
```
