```python
from pyspark.sql import functions as sparkf

raw_df = spark.createDataFrame(
 [
  ('Store 1',1,448),
  ('Store 1',2,None),
  ('Store 1',3,499),
  ('Store 1',44,432),
  (None,None,None),
  ('Store 2',1,355),
  ('Store 2',1,355),
  ('Store 2',None,345),
  ('Store 2',3,387),
  ('Store 2',4,312)
],
 ['Store','WeekInMonth','Revenue']
)
```

```python
raw_df.count()
```

```python
raw_df.printSchema()
```

```python
raw_df.select(['Store']).show()
```

```python
raw_df.select([col for col in raw_df.columns]).show()
```

```python
raw_df.select([col for col in raw_df.columns if col.startswith('Rev')]).show()
```

```python
raw_df.select([sparkf.count(col) for col in raw_df.columns if col.startswith('Rev')]).show()
```

```python
raw_df.select([sparkf.count(col) for col in raw_df.columns]).show()
```

```python
raw_df.select([sparkf.count(sparkf.when(sparkf.col(col).isNull(),col)) for col in raw_df.columns]).show()
```

```python
raw_df.select([sparkf.count(sparkf.when(sparkf.col(col).isNull(),col)).alias(col) for col in raw_df.columns]).show()
```
